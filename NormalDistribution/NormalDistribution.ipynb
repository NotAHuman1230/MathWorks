Format and setup the data for easier manipulation

```python
import pandas as pd

rawData = pd.read_csv("Data.csv")
#These are the values most likely to change
xField = "Height"
types = ["Male", "Female"]

for i in rawData.columns[1::]:
    rawData[i] = pd.to_numeric(rawData[i], errors="coerce")

rawData = rawData.dropna()

data = pd.DataFrame()
for i in types:
    data[i] = rawData[rawData["Gender"] == i][xField].tolist()
data.head()
```

Calculate the means and standard deviations to draw the normal distribution

```python
means = [data[i].mean() for i in types]
std = [data[i].std() for i in types]

print(f"Means: {means}")
print(f"STD: {std}")
```

Plot the normal distribution with the actual data

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plotting():
    #These are the values most likely to change
    plt.title("Heights of males and females")
    plt.xlabel("Height")
    plt.ylabel("Frequency Density") 

    plt.legend(loc="upper left", prop={'size': 8})

binAmount = 100

xValues = np.linspace(min(data.min()) * 0.95, max(data.max()) * 1.05, 10000)
for i in range(len(types)):
    plt.plot(xValues, stats.norm.pdf(xValues, means[i], std[i]), label=types[i] + " Distribution")
    plt.hist(data[types[i]], density=True, label=types[i] + " Histogram", bins=binAmount, alpha=0.5)

plotting()
```
